```python
from sklearn.datasets import load_diabetes #https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html
import numpy as np
import os
```

```python
output_path = "NAI_04/data/diabetes_dataset.txt"

data = load_diabetes(as_frame=False)

X = data.data
y = data.target
```

```python
full_data = np.hstack([X, y.reshape(-1, 1)])

os.makedirs("data", exist_ok=True)

np.savetxt(output_path, full_data, fmt="%.5f", delimiter="\t")

print(f"Dataset zapisano w: {output_path}")
```
